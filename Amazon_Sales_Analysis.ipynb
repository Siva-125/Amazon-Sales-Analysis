## **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Data Extracting**

```python
data = pd.read_csv('/content/sample_data/amazon.csv')
data.head()
```

```python
data.info()
```

```python
data.describe().T
```

## **Data Cleaning**

```python
data.duplicated().sum()
data.drop_duplicates(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.nunique()
```

```python
data.columns
```

```python
data = data.drop(columns=['product_id','user_id', 'user_name', 'review_id', 'img_link', 'product_link'])
data.head()
```

```python
data['discounted_price'] = data['discounted_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float).astype(int)
data['actual_price'] = data['actual_price'].str.replace('₹', '', regex=False).str.replace(',', '', regex=False).astype(float).astype(int)
data['rating_count'] = data['rating_count'].str.replace(',', '', regex=False).astype(float).astype(int)
data['rating'] = data['rating'].str.replace('|', '', regex=False).replace('', np.nan).astype(float)
data['discount_percentage'] = data['discount_percentage'].str.replace('%', '', regex=False).astype(float).astype(int)

data.info()
```

```python
data.head()
```

```python
prices_corr = data[['actual_price', 'discounted_price', 'discount_percentage', 'rating', 'rating_count']].corr()
sns.heatmap(prices_corr, annot=True)
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

## **Exploratory Data Analysis**

```python
# Most Expensive products
data.groupby('product_name')['discounted_price'].max().sort_values(ascending=False).head(10)
```

Most of the expensive products are TVs

```python
# Average ratings for each products
avg_ratings = data.groupby('product_name')['rating'].mean()
avg_ratings
```

```python
# Top 10 High Rated Products
top_10_rated = data.groupby('product_name')['rating'].mean().sort_values(ascending=False).head(10)
top_10_rated
```

Amazon Wireless Mouse and Syncwire Cable gets 5/5 rating among other products

```python
# Which category products gets best sales?
data.groupby('category')['product_name'].count().sort_values(ascending=False).head(5)
```

Computer & Accessories, Peripherals, Cables, USB Cables categories are solds more.

```python
category_sales = data.groupby('category')['product_name'].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
plt.title('Category Sales Distribution')
plt.axis('equal')
```

```python
# Which category products get worst sales?
data.groupby('category')['product_name'].count().sort_values().head(10)
```

Products related to hardware, networking devices, vehcle accessories are solds less.

```python
plt.title('Actual Price vs Discounted Price')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
sns.scatterplot(data=data, x='actual_price', y='discounted_price')
plt.show()
```

From Above Scatter Plot, when the actual price is more then the discount is also increases.

### **Key Insights from Amazon Sales**

- High priced products are dominated TVs, indicateing that large electronics are sold at higher price
- Amazon Wireless Mouse and Syncwire Cable both products recieved a perfect 5/5 ratings.
- Top Selling Categories:
  - Computer & Accessories
  - Peripherals
  - Cables
- Least Selling Categories:
  - Hardware components
  - Networking devices
  - Vehicle Accessories
- Price vs. Discount Relationship: From the scatter plot, it is clear that higher-priced products tend to have larger discounts, suggesting a pricing strategy to attract buyers for expensive items.
